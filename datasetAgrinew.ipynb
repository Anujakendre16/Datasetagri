```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy


file_path = "databaseAgri.csv"
data = pd.read_csv(file_path)

print(data.head())
```

```python
print(data.info())
print(data.describe())
print(data.isnull().sum())
```

```python

simplified_data = data[[
    "Plant Height (PH)", 
    "Number of Pods (NP)", 
    "ChlorophyllA663", 
    "Protein Percentage (PPE)", 
    "Seed Yield per Unit Area (SYUA)"   # Target variable
]]
```

```python
simplified_data.to_csv("simplified_agri_dataset.csv", index=False)
print("New simplified dataset saved!")
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Height vs Yield
plt.scatter(simplified_data["Plant Height (PH)"], simplified_data["Seed Yield per Unit Area (SYUA)"])
plt.xlabel("Plant Height (PH)")
plt.ylabel("Seed Yield (SYUA)")
plt.show()
```

```python
correlation_matrix = simplified_data.corr()
print(correlation_matrix)
```

```python
plt.scatter(simplified_data["Plant Height (PH)"], simplified_data["Seed Yield per Unit Area (SYUA)"])
sns.set_style("dark")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=simplified_data, x="Seed Yield per Unit Area (SYUA)", kde=True)

# Add labels and title
plt.xlabel("Seed Yield per Unit Area (SYUA)")
plt.ylabel("Frequency")
plt.title("Distribution of Seed Yield")

# Show the plot
plt.show()

```
